 NORMAL DISTRIBUTION (using mean and standard deviation)

```python
from IPython.core.interactiveshell import py3compat
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


mean_val=0.5
standard_dev=0.1

np.random.seed(0)

X=np.random.normal(mean_val,standard_dev,(395,1))
Y=np.random.normal(mean_val*2,standard_dev*3,(395,1))
 
plt.scatter(X,Y,color='g')
plt.show() 
```
